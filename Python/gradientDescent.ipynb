```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Dataset File.csv")
X = data.iloc[:,0]
y = data.iloc[:,1]
```

```python
function = lambda x: 2*x-1
```

```python
function(5)
```

```python
X = np.linspace(1,10,10)
X.shape
```

```python
noise = np.random.normal(0,1,10)
noise
```

```python
y = function(X)+noise
y
```

```python
plt.scatter(X, y)
```

```python
w = np.random.random()
b = np.random.random()
```

```python
w,b
```

```python
l = 0.0001
epochs = 100000

n = len(X)

for i in range(epochs):
  Y = w*X + b
  
  dw = (-2/n) * sum(X * (y - Y))
  db = (-2/n) * sum(y - Y)

  w -= l*dw
  b -= l*db

print(w,b)
```

```python
plt.scatter(X, y)
plt.plot(X, Y, color='orange')
plt.show()
```
