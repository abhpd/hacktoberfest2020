```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = x**2
```

```python
print(x,y)
```

```python
plt.plot(x,y,label="Square",color='red',marker='o')
plt.plot(x,x**3,label="Cube",color='green', linestyle='dashed')
plt.title('Line Plot')
plt.xlabel("Numbers")
plt.ylabel("Square/Cube")
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-dark')
```

Scatter Plot

```python
plt.scatter(x,y,label="Square")
plt.scatter(x,x**3,label="Cube")
plt.title('Scatter Plot')
plt.xlabel("Numbers")
plt.ylabel("Square/Cube")
plt.legend()
plt.show()
```

```python
plt.style.available
```

Bar Graph

```python
plt.style.use('dark_background')
xcord = np.array([0,1,2])*2
plt.bar(xcord,[10,20,15],width=0.5,label='Data1',tick_label=['one','two','three'],color='green') #Data 1
plt.bar(xcord+0.5,[20,10,17],width=0.5,label='Data2',color='blue') #Data 2
plt.ylim(0,25)
plt.legend()
plt.show()
```

Pie Chart

```python
plt.style.use('seaborn')
subjects= 'Maths','Physics','Chemistry','English'
weightage = [20,10,15,5]
plt.pie(weightage,labels=subjects,explode=(0,0.1,0,0),autopct='%1.1f%%',shadow=True)
plt.show()
```

Histogram

```python
Xsn = np.random.randn(100)
sigma = 8
u = 70
x = np.round(Xsn*sigma+u)
x2 = np.round(Xsn*5+40)
```

```python
x
```

```python
plt.style.use('seaborn')
plt.hist(x,alpha=0.9)
plt.hist(x2,alpha=0.8)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

Multivarience Normal Distribution

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.show()
```
