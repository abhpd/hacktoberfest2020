```python
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

theta, phi = np.mgrid[0:2 * np.pi:30j, 0:2*np.pi:50j]

a, b, c, r = 2, 2, 2, 2

x = a * (r + np.sin(phi)) * np.cos(theta)
y = b * (r + np.sin(phi)) * np.sin(theta)
z = c * np.cos(phi)

fig = plt.figure(figsize=(6, 24), dpi=130)

ax1 = fig.add_subplot(411, projection='3d')
ax1.set_xlim(-6, 6)
ax1.set_ylim(-6, 6)
ax1.set_zlim(-6, 6)
ax1.set_box_aspect((1, 1, 1))
ax1.plot_surface(x, y, z, cmap=cm.autumn)
ax1.view_init(90, 0)
ax1.set_ylabel('x-y view')

ax2 = fig.add_subplot(412, projection='3d')
ax2.set_xlim(-6, 6)
ax2.set_ylim(-6, 6)
ax2.set_zlim(-6, 6)
ax2.set_box_aspect((1, 1, 1))
ax2.plot_surface(x, y, z, cmap=cm.autumn)
ax2.view_init(0, 90)
ax2.set_xlabel('y-z view')

ax3 = fig.add_subplot(413, projection='3d')
ax3.set_xlim(-6, 6)
ax3.set_ylim(-6, 6)
ax3.set_zlim(-6, 6)
ax3.set_box_aspect((1, 1, 1))
ax3.plot_surface(x, y, z, cmap=cm.autumn)
ax3.view_init(0, 0)
ax3.set_ylabel('x-z View')

ax = fig.add_subplot(414, projection='3d')
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_zlim(-6, 6)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
ax.set_box_aspect((1, 1, 1))

ax.plot_surface(x, y, z, cmap=cm.autumn)
plt.show()
```
