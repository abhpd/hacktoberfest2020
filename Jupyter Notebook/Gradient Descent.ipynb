```python
#This is a simple program to demonstrate Gradient descent algorithm
import numpy as np
```

```python
def gradientdescent(x,y):
    m=b=0
    itr=1000
    n= len(x)
    lr=0.0002
    
    for i in range(itr):
        y_predicted=(m*x + b)
        cost= (1/n) * sum([val**2 for val in (y-y_predicted)])
        m_derivative= -(2/n)*sum(x*(y-y_predicted))
        b_derivative= -(2/n)*sum(y-y_predicted)
        m= m- lr*m_derivative
        b= b- lr*b_derivative
        print("m={} , b={} ,cost={} , itr={}".format(m,b,cost,i))

x= np.array([92,56,88,70,80,49,65,35,66,67])
y= np.array([98,68,81,80,83,52,66,30,68,73])

gradientdescent(x,y)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.xlabel('X- axis')
plt.ylabel('Y-axis')
plt.show()
```
